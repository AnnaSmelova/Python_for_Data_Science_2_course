## Решение домашнего задания к уроку 2 “Анализ данных и проверка статистических гипотез.”

### Задание 1
#### В чём различие между зависимыми и независимыми выборками?

Выборки называются **независимыми (несвязными)**, если процедура эксперимента и полученные результаты измерения некоторого свойства у испытуемых одной выборки не оказывают влияния на особенности протекания этого же эксперимента и результаты измерения этого же свойства у испытуемых (респондентов) другой выборки.

И, напротив, выборки называется **зависимыми (связными)** если процедура эксперимента и полученные результаты измерения некоторого свойства, проведенные на одной выборке, оказывают влияние на другую. Одна и та же группа испытуемых, на которой дважды проводилось психологическое обследование, по определению оказывается зависимой, или связной выборкой.

Примеры:
1. пары близнецов - зависисмые
2. два измерения какого-либо признака до и после экспериментального воздействия - зависимые
3. мужья и жёны - зависимые
4. мужчины и женщины - независисмые
5. психологи и математики - независимые

Своими словами: если процесс получения первой выборки не влияет на получение второй выборки, то такие выборки называются независимыми. Если влияет, то зависимыми.

### Задание 2
#### Когда применяются параметрические статистические критерии, а когда — их непараметрические аналоги?

**Параметрические** методы основаны на некоторых, как правило, вполне вероятных предположениях о характере распределения случайной величины. Обычно параметрические методы, используемые в анализе экспериментальных данных, основаны на предположении нормальности распределения этих данных. Следствием такого предположения является необходимость оценки исследуемых параметров распределения. 

**Непараметрическими** называют методы математической статистики, при которых не выдвигаются какие-либо априорные предположения о характере распределения исследуемых данных и не предполагается каких-либо допущений о соотношении параметров распределения анализируемых величин. В этом заключается главное достоинство этих методов.

Источник: https://studme.org/50296/psihologiya/parametricheskie_neparametricheskie_protsedury_statisticheskogo_analiza_dannyh

Своими словами: параметрические критерии применяются, когда знаем закон распределения признака(например, распределение является нормальным), а непараметрические, когда не знаем.
