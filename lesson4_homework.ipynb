### Задание 1
#### Расскажите, как работает регуляризация в решающих деревьях, какие параметры мы штрафуем в данных алгоритмах?

Регуляризация - метод введения штрафов для модели целью снижения переобучения. В качестве регуляризации в решающих деревьях используются критерии останова, которые показывают, нужно ли остановить процесс построения дерева. Правильный выбор критериев останова роста дерева может существенно повлиять на его качество. Существует большое количество возможных ограничений, например:
- Ограничение максимальной глубины дерева
- Ограничение максимального количества листьев
- Ограничение минимального количества объектов в листе
- Останов в случае, когда все объекты в листе относятся к одному классу

Альтернативой критериям останова является стрижка деревьев. При использовании стрижки сначала строится переобученное дерево (например, до тех пор, пока в каждом листе не окажется по одному объекту), а затем производится оптимизация его структуры с целью улучшения обобщающей способности. Существует ряд исследований, показывающих, что стрижка позволяет достичь лучшего качества по сравнению с ранним остановом построения дерева на основе различных критериев. Тем не менее, на данный момент методы стрижки редко используются и не
реализованы в большинстве библиотек для анализа данных. Причина заключается в том, что деревья сами по себе являются слабыми алгоритмами и не представляют большого интереса, а при использовании в композициях они либо должны быть переобучены (в случайных лесах), либо должны иметь очень небольшую глубину (в бустинге), из-за чего необходимость в стрижке отпадает.

### Задание 2
#### По какому принципу рассчитывается "важность признака (feature_importance)" в ансамблях деревьев?

На каждом шаге построения дерева выбирается признак и его значение, которые дают наилучшее разбиение данных. В зависимости от алгоритма, применяются разные способы подсчета прироста информации, чтобы вычислить наилучшее разбиение(например считают критерий Джини или энтропию в каждой части разбиения). Важность признака в дереве  определяется степенью снижения меры неопределенности, т.е. как раз критерия Джини или энтропии и других. Чем выше признак в дереве, тем он важнее. <br>
В ансамблях деревьев важность признака считается как среднее арифметическое по всем важностям данного признака для каждого дерева. В случае, если признак не присутствует в дереве, его важность считается нулевой.
